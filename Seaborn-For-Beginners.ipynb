# **1.Distribution plot**

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.displot(data=penguins, x="flipper_length_mm")
```

```python
# Use the kind parameter to select a different representation:
sns.displot(data=penguins, x="flipper_length_mm", kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="ecdf")
```

```python
# While in histogram mode, it is also possible to add a KDE curve:
sns.displot(data=penguins, x="flipper_length_mm", kde=True)
```

```python
# bivariate means two values or two variables here !!!

#To draw a bivariate plot, assign both x and y

sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm")
```

```python
# Rug-plot 
tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip", height=-.02, clip_on=False)
```

#  **2.Categorical plot**

```python
# Bar plot

tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
```

### Estimator parameter

```python
import numpy as np 
# try with np.sum and np.mea
sns.barplot(x="day", y="total_bill", hue="sex", data=tips,estimator=np.std)
```

```python
sns.barplot(x="day", y="tip", data=tips, capsize=.2)
```

```python
# Countplot 
titanic = sns.load_dataset("titanic")
sns.countplot(x="class", data=titanic)
```

```python
sns.countplot(x="class", hue="who", data=titanic) # added hue parameter 
```

```python
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# strip plot
#Group the strips by a categorical variable:
tips = sns.load_dataset("tips")
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="sex", y="total_bill", hue="day", data=tips)
```

```python
# swarmplot
# Draw a categorical scatterplot with non-overlapping points.

sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
# added hue args
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
sns.swarmplot(x="day", y="total_bill", data=tips,color="white", edgecolor="gray")
```

# 3. **Matrix plot**

```python
flights = sns.load_dataset("flights")
fls_cor = flights.corr()
sns.heatmap(fls_cor)
```

```python
sns.heatmap(fls_cor,annot=True,cmap='coolwarm') 
```

```python
flights = flights.pivot("month", "year", "passengers")
```

```python
#here index and col not in order , they are clustring(grouping) the common data
sns.clustermap(flights,cmap='mako') 
```

# 4. **Grids**

```python
# Pair plot 
# Plot pairwise relationships in a dataset.
penguins = sns.load_dataset("penguins")
sns.pairplot(penguins)
```

```python
# added categorical hue args
sns.pairplot(penguins, hue="species")
```

```python
sns.pairplot(penguins, hue="species", diag_kind="hist")
```

```python
# FacetGrid
# Multi-plot grid for plotting conditional relationships.

tips = sns.load_dataset("tips")
sns.FacetGrid(tips)
```

```python
# Assign column and/or row variables to add more subplots to the figure:
sns.FacetGrid(tips, col="time", row="sex")
```

```python
obj = sns.FacetGrid(tips, col="time",  row="sex")
obj.map(sns.scatterplot, "total_bill", "tip")
```

```python
obj = sns.FacetGrid(tips, col="time", hue="sex")
obj.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
obj.set_axis_labels("Total bill", "Tip")
obj.add_legend()
```

# 5. **Regression plot**

```python
# lmplot  Linear model plot
import seaborn as sns; sns

tips = sns.load_dataset("tips")
g = sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
 sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,markers=["o", "x"])
```

```python
sns.lmplot(x="size", y="total_bill", hue="day", col="day",data=tips, height=6, aspect=.4, x_jitter=.1)
```

```python
# Reg plot regression plot
tips = sns.load_dataset("tips")
ax = sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
# Fit a higher-order polynomial regression:

ans = sns.load_dataset("anscombe")

ax = sns.regplot(x="x", y="y", data=ans.loc[ans.dataset == "II"],
                 scatter_kws={"s": 80},
                 order=2, ci=None)
```

```python
# Fit the regression model using log(x) function : 
sns.regplot(x="size", y="total_bill", data=tips,
                 x_estimator=np.mean, logx=True)
```
